```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

def plot_histograms_boxplots(data, column, group_col, group1, group2):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    group1_data = data[data[group_col] == group1][column]
    group2_data = data[data[group_col] == group2][column]

    sns.histplot(group1_data, kde=True, color="blue", label=group1, ax=axes[0], alpha=0.5)
    sns.histplot(group2_data, kde=True, color="red", label=group2, ax=axes[0], alpha=0.5)
    axes[0].set_title(f'Histogram of {column} ({group1} vs {group2})')
    axes[0].legend()

    sns.boxplot(x=data[group_col], y=data[column], ax=axes[1])
    axes[1].set_title(f'Boxplot of {column} by {group_col}')

    plt.tight_layout()
    plt.show()

plot_histograms_boxplots(df, column="DV", group_col="SECTION", group1="ALPHA", group2="BETA")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

p = pd.DataFrame({
    'DV': np.random.choice(['A', 'B', 'C'], size=718),  # Categorical variable
    'Category': np.random.choice(['X', 'Y'], size=718)  # Another categorical variable
})

# 🔹 Step 1: Convert 'DV' column to categorical if needed
p['DV'] = p['DV'].astype(str)  # Ensure it's a string (not mixed types)

# 🔹 Step 2: Create a Contingency Table
contingency_table = pd.crosstab(p['Category'], p['DV'])

# 🔹 Step 3: Convert to NumPy array (2D) for chi2_contingency
contingency_array = contingency_table.values

# 🔹 Step 4: Perform the Chi-Square Test
stat, p_value, dof, expected = chi2_contingency(contingency_array)

# 🔹 Step 5: Set the significance level
alpha = 0.05

# 🔹 Step 6: Print results
print("Contingency Table:\n", contingency_table)
print("\nChi-Square Statistic:", stat)
print("p-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# 🔹 Step 7: Decision Making
if p_value <= alpha:
    print("\n❌ The variables are dependent (Reject H0)")
else:
    print("\n✅ The variables are independent (Fail to reject H0)")
```
